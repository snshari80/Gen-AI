## Q1 — List + Dictionary: Student Marks Tracker

You are building a simple marks tracker for a class of students.

**Tasks:**
1. Create a **list** of 5 student names of your choice
2. Create a **dictionary** that maps each student name to a list of 3 marks (out of 100) — one mark per subject: `Maths`, `Science`, `English`
3. Loop through the dictionary and print each student's **name** and their **average mark** (rounded to 1 decimal place)
4. Find and print the student with the **highest average**
5. Add a **new student** with 3 marks to the dictionary and print the updated count of students

**Expected output format:**
```
Arjun   → Average: 87.3
Priya   → Average: 91.0
...
Top student: Priya with 91.0
Total students after addition: 6
```

````python
students = [
    {
        'name': 'John',
        'Maths': 95,
        'Science': 88,
        'English': 73
    },
    {
        'name': 'Harish',
        'Maths': 96,
        'Science': 98,
        'English': 86
    },
    {
        'name': 'Ravi',
        'Maths': 85,
        'Science': 78,
        'English': 66
    },
        {
        'name': 'Joel',
        'Maths': 75,
        'Science': 98,
        'English': 86
    },
        {
        'name': 'Kumar',
        'Maths': 55,
        'Science': 98,
        'English': 56
    }
]
subjects = ['Maths', 'Science', 'English']
max_score = {}
for i in students:
    avg = sum(i[subject] for subject in subjects) / len(subjects)
    max_score.update({i['name']: avg})
    print(f"{i['name']} -> Average: {avg:.1f}")
max_score = {k: v for k, v in max_score.items() if v == max(max_score.values())}
for i in max_score:
    print(f"Top Student: {i} with {max_score[i]:.1f}")
students.append({
    'name': 'Rohit',
    'Maths': 85,
    'Science': 78,
    'English': 66
})
print("Total students after addition:", len(students))
````

## Q3 — Stack & Queue: Customer Support System

A customer support system handles two types of tickets:
- **VIP tickets** are handled with a **Stack** (most recent VIP issue handled first — LIFO)
- **Regular tickets** are handled with a **Queue** (first come, first served — FIFO)

**Tasks:**
1. Create an empty **list** called `vip_stack` and an empty **deque** called `regular_queue`
2. Add these VIP tickets to the stack (in this order): `'VIP-001'`, `'VIP-002'`, `'VIP-003'`
3. Add these regular tickets to the queue (in this order): `'REG-101'`, `'REG-102'`, `'REG-103'`
4. Process (remove and print) the **top VIP ticket** from the stack
5. Process (remove and print) the **first regular ticket** from the queue
6. Add two new tickets: `'VIP-004'` to the stack and `'REG-104'` to the queue
7. Print the **current state** of both the stack and queue after all operations

**Expected output format:**
```
Processing VIP ticket : VIP-003
Processing REG ticket : REG-101
VIP stack now   : ['VIP-001', 'VIP-002', 'VIP-004']
Regular queue now: deque(['REG-102', 'REG-103', 'REG-104'])
```

````python
from collections import deque
vip_stack = []
regular_queue = deque()

for i in ['VIP-001', 'VIP-002', 'VIP-003']:
    vip_stack.append(i)

for i in ['REG-101', 'REG-102', 'REG-103']:
    regular_queue.append(i)

stack_ticket = vip_stack.pop()
print(f"Processed VIP ticket: {stack_ticket}")
regular_queue_ticket = regular_queue.popleft()
print(f"Processed REG ticket: {regular_queue_ticket}")
vip_stack.append('VIP-004')
regular_queue.append('REG-104')
print(f"VIP stack now: {vip_stack}")
print(f"Regular queue now: {list(regular_queue)}")
````

## Q4 — Tuple, NamedTuple & Frozenset: Product Catalogue

You are building a product catalogue for an online store.

**Tasks:**
1. Create a **NamedTuple** called `Product` with fields: `name`, `category`, `price`, `in_stock` (True/False)
2. Create **4 product instances** using this NamedTuple — at least 2 different categories (e.g. Electronics, Clothing)
3. Print each product's **name and price** by accessing fields by name (not index)
4. Create a **frozenset** of categories from your products (use a set comprehension, then freeze it)
5. Use the frozenset as a **dictionary key** that maps to the count of products in that category — print the result
6. Filter and print only products that are **in stock** and cost **less than ₹5000**

**Expected output format:**
```
Laptop      → ₹45,000
T-Shirt     → ₹799
...
Categories (frozenset): frozenset({'Electronics', 'Clothing'})
In stock under ₹5000  : ['T-Shirt', ...]
```

````python
from collections import namedtuple

Product = namedtuple('Product', ['name', 'in_stock', 'category', 'price'])
product_category = frozenset()
product_count_based_category = {}

prod1 = Product(name='Laptop', in_stock=True, category='Electronics', price=35678)
prod2 = Product(name='Book', in_stock=False, category='Education', price=1856)
prod3 = Product(name='T-shirt', in_stock=True, category='Clothing', price=1458)
prod4 = Product(name='Smartphone', in_stock=False, category='Electronics', price=25789)

for i in [prod1, prod2, prod3, prod4]:
   product_category = product_category | {i.category}
   print(f"{i.name} -> ₹{i.price:,}")
for item in product_category:
   count = len([p for p in [prod1, prod2, prod3, prod4] if p.category == item])
   product_count_based_category.update({item: count})

print(f"Categories (frozenset): {product_category}")
print(f"Product count by category: {product_count_based_category}")
print(f"In Stock and Under ₹50,000: { [p for p in [prod1, prod2, prod3, prod4] if p.in_stock == True and p.price <= 50000] }")
````

## Q5 — Mixed Structures: Library Book Management

This question combines multiple data structures to build a mini library system.

**Tasks:**
1. Create a **dictionary** called `library` where:
   - Each **key** is a genre (string): `'Fiction'`, `'Science'`, `'History'`
   - Each **value** is a **list** of book titles (at least 3 books per genre)
2. Create a **set** called `borrowed_books` containing at least 2 book titles that are currently borrowed
3. For each genre, print the genre name and the books that are **available** (not in `borrowed_books`)
4. A user wants to borrow `'A Brief History of Time'` — check if it exists in the library (any genre) and if it is not already borrowed. Print an appropriate message.
5. Add the borrowed book to `borrowed_books`, then print the **total number of books available** across all genres (total books minus borrowed books count)
6. Create a **tuple** that stores a summary: `(total_genres, total_books, total_borrowed)` and print it

**Expected output format:**
```
Fiction  → Available: ['The Alchemist', 'To Kill a Mockingbird']
Science  → Available: ['Cosmos', ...]
...
'A Brief History of Time' is available — borrowing now.
Total available books : 7
Summary tuple: (3, 9, 3)
```

````python
library_books = {
    'Fiction': ['The Great Gatsby', 'To Kill a Mockingbird', '1984'],
    'Science': ['A Brief History of Time', 'The Selfish Gene', 'The Origin of Species'],
    'History': ['Sapiens', 'Guns, Germs, and Steel', 'The Silk Roads']
}

borrowed_books = {
    'The Great Gatsby', 'The Origin of Species', 'The Silk Roads',
}

book = library_books['Science'][0]

for k, i in library_books.items():
    print(f'{k} => Available: {[book for book in i if book not in borrowed_books]}')

print(f"'{book}' is available - borrowing it now!"
      if book not in borrowed_books 
      else f"'{book}' is borrowed")
borrowed_books.add(book)

print(f'Updated borrowed books: {borrowed_books}')
print(f'Total available books: {sum(len([book for book in i if book not in borrowed_books]) for i in library_books.values())}')
print(f"Total borrowed books: {len(borrowed_books)}")
print(f"Total categories: {len(library_books.keys())}")

total_books = sum(len(books) for books in library_books.values())
summary = (len(library_books.keys()), total_books, len(borrowed_books))

print(f"Summary of library books Tuple: {summary}")
````
