```python
import requests
import json
import pandas as pd

print('All Imported successfully')

base_url='https://dummyjson.com'

all_post = []

def fetch_post(url:str)->list:
    full_path = f'{base_url}/{url}'
    print(full_path,"full_path")
    response = requests.get(full_path)
    if(response.status_code != 200):
        print('Error Accured')
        return ''
    posts = response.json().get('posts',[])
    return posts

all_post = fetch_post('posts')

print(json.dumps(all_post,indent=2))


def create_pddataframe(allpost:list) -> pd.DataFrame:
    rows = []
    for i in allpost:
        rows.append({
            "title": i.get('title') or '',
            'userId': i.get('userId') or '',
            'Likes':i.get('reactions',{}).get('likes'),
            'Dislikes':i.get('reactions',{}).get('dislikes'),
            'Content': " ".join(filter(None,[i.get('title'),i.get('body')]))
            
        })
    df = pd.DataFrame(rows)
    return df


create_pddataframe(all_post)
```
