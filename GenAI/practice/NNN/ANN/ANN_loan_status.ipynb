Loading the data

```python
import pandas as pd

df = pd.read_csv('loan_clean_dataset.csv')

print(df.head())
df.describe()
```

EDA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

df.describe()
df.isnull().sum()


X = df.drop(columns=['Approved']).values
y = df['Approved'].values

scalar = StandardScaler()


X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.2,random_state=7
)

X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)

print(f'Length of test: {len(X_test)}')
print(f'Length of train: {len(X_train)}')
```

Model Training

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer

model = Sequential([
    InputLayer(input_shape=(5,)),
    Dense(12, activation="relu"),
    Dense(8, activation="relu"),
    Dense(4, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.summary()

model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"],
)

histroy = model.fit(
    X_train,y_train, epochs = 30, batch_size = 5,   validation_split=0.2
)

print("\nTraining complete!")

training_loss, training_accuracy = model.evaluate(X_train,y_train)
test_loss, test_accuracy = model.evaluate(X_test,y_test)

print(f'Training Accuracy Rate: {training_accuracy *100:.2f}%')
print(f'Training Loss: {training_loss}')

print(f'Testing Accuracy Rate: {test_accuracy *100:.2f}%')
print(f'Testing Loss: {test_loss}')
```

Predict on test data


```python
prediction = model.predict(X_test)

X_org = scalar.inverse_transform(X_test)

header = f"{'Age':>5} {'Income':>8} {'LoanAmount':>12} {'CreditScore':>12} {'Employed':>10} {'Actual':>15} {'Predicted':>15}"
print(header)
print("-" * len(header))

for i in range(5):
  row = X_org[i]
  prob = prediction[i][0]
  Predicted = "Approved ✅" if prob >= 0.5 else "Rejected ❌"
  Actual = "Approved ✅" if y_test[i]==1 else "Rejected ❌"
  print(f"{row[0]:>5.0f} {row[1]:>8.0f} {row[2]:>12.0f} {row[3]:>12.0f} {row[4]:>10.0f} {Actual:>15} {Predicted:>15}")
```
